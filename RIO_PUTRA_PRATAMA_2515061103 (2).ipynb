**VECTOR DAN VEKTOR TRANSPOSISI**

```python
import numpy as np
```

```python
# Soal 7.1 - Buat vector x dengan ukuran 3x1
x = np.array([103, 10, 3]) # type argument is optional, e.g.: dtype=np.float16
x
```

```python
# Hitung panjang vector x
len(x)
```

```python
# Tampilkan ukuran vector x
x.shape
```

```python
# Tampilkan tipe data dari vector x
type(x)
```

```python
# Akses elemen pertama dari vector x
x[0] # zero-indexed
```

```python
# Tampilkan tipe data elemen pertama dari vector x
type(x[0])
```

```python
# Soal 7.2 - Lakukan transpose pada vector x dan simpan pada variabel x_t
# Transposing a regular 1-D array has no effect...
x_t = x.T
x_t
```

```python
# Tampilkan ukuran dari vector x_t
x_t.shape
```

```python
# Lakukan transpose pada vector y dan simpan pada variabel y_t
# ...but it does we use nested "matrix-style" brackets:
y = np.array([[103, 10, 3]])
y
```

```python
y.shape
```

```python
# ...but can transpose a matrix with a dimension of length 1, which is mathematically equivalent:
y_t = y.T
y_t
```

```python
# Tampilkan ukuran dari vector y_t
y_t.shape # this is a column vector as it has 3 rows and 1 column
```

```python
# Column vector dapat di-transpose kembali ke row vector semula:
y_t.T
```

```python
y_t.T.shape
```

```python
# Soal 7.3 - Lakukan soal nomor 2 menggunakan pytorch dan tensorflow
import torch
import tensorflow as tf
```

```python
# Buat vector x menggunakan pytorch
x_pt = torch.tensor([103, 10, 3])
x_pt
```

```python
# Transpose vector x menggunakan pytorch
x_pt_t = x_pt.reshape(1, -1).T
x_pt_t
```

```python
x_pt_t.shape
```

```python
# Buat vector x menggunakan tensorflow
x_tf = tf.Variable([25, 2, 5])
x_tf
```

```python
# Transpose vector x menggunakan tensorflow
x_tf_t = tf.transpose(tf.expand_dims(x_tf, 0))
x_tf_t
```

```python
x_tf_t.shape
```

**NORMS AND UNIT VECTORS (L2 NORMS)**

```python
import numpy as np
```

```python
# Soal 8.1 - Buat vector x dengan ukuran 3x1
x = np.array([103, 10, 3]) # type argument is optional, e.g.: dtype=np.float16
x
```

```python
# Hitung nilai L2 norm dari vector x
np.linalg.norm(x)
```

```python
# Soal 8.2 - Buat vector v dengan elemen (3, 4)
v = np.array([3, 4])
v
```

```python
# Hitung nilai L2 norm dari vector v
np.linalg.norm(v)
```

```python
# Tampilkan unit vector dari vector v
v_unit = v / np.linalg.norm(v)
v_unit
```

```python
# Hitung panjang unit vector dari vector v
np.linalg.norm(v_unit)
```

**BASIS, ORTHOGONAL, AND ORTHONORMAL VECTORS**

```python
import numpy as np
```

```python
# Soal 9.1 - Buat vector i dengan elemen (1,0) dan vector j dengan elemen (0,1)
i = np.array([1, 0])
i
```

```python
j = np.array([0, 1])
j
```

```python
# Buktikan bahwa vector i dan vector j itu orthogonal
np.dot(i, j) # detail on the dot operation coming up...
```

```python
# Soal 9.2 - Bagaimana dengan v=(1,2) dan w=(2,-1), apakah juga orthogonal?
v = np.array([1, 2])
v
```

```python
w = np.array([2, -1])
w
```

```python
np.dot(v, w) # jika hasil dot product = 0, maka vector v dan w orthogonal
```
